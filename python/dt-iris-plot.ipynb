```python
from sklearn.datasets import load_iris

iris = load_iris()
X, y = iris.data, iris.target

X, y
```

```python
import numpy as np
import pandas as pd

irisdf = pd.DataFrame(
    np.hstack((X, y[:, np.newaxis])),
    columns=['sepal-length', 'sepal-width',
             'petal-length', 'petal-width', 'class']
)
kind_dict = {0: "setosa", 1: "versicolor", 2: "virginica"}
irisdf["class"] = irisdf["class"].map(kind_dict)

irisdf.head()
```

```python
irisdf.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

with plt.style.context('Solarize_Light2'):
    g = sns.pairplot(irisdf, diag_kind="kde", kind="scatter", hue="class", markers=["o", "s", "D"])
    plt.show()
```

```python
with plt.style.context('Solarize_Light2'):
    g = sns.pairplot(irisdf, diag_kind="hist", kind="kde", hue="class")
    plt.show()
```

```python
from sklearn import tree
import graphviz 

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True, special_characters=True) 
graph = graphviz.Source(dot_data) 
graph
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4)
clf = clf.fit(X, y)
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True, special_characters=True) 
graph = graphviz.Source(dot_data) 
graph
```
